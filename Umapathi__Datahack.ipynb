```python
import numpy as np
import pandas as pd
import sklearn
```

```python
# loading the training feature csv file
df1=pd.read_csv('/content/training_set_features.csv')
df1
```

```python
# loading the training label dataframe
df2=pd.read_csv('/content/training_set_labels.csv')
# merging training set features csv with training set labels
df=pd.merge(df1,df2,on='respondent_id')
df
```

```python
# replacing NaN values with 0
df.fillna(0)
```

```python
# doing feature engineering
pd.crosstab(df['xyz_concern'],df['seasonal_vaccine'])
```

```python
# feature engineering
# analyzing the different features to judge the importance of the features\\
# plotting
import plotly.express as px
import seaborn as sns
#sns.countplot(data=df[df['xyz_concern']],y='seasonal_vaccine')
fig=px.histogram(df,x='seasonal_vaccine',color='xyz_concern',barmode='group')
fig.show()
```

```python
# displaying the features name which are columns
df.columns
```

```python
# removing less important features in classification
columns_to_drop=[
       'behavioral_antiviral_meds',
       'behavioral_face_mask', 'behavioral_wash_hands',
        'behavioral_outside_home',
       'behavioral_touch_face',
       'health_insurance', 'age_group',
      'education', 'race', 'sex', 'income_poverty', 'marital_status',
       'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa',
       'household_adults', 'household_children', 'employment_industry',
       'employment_occupation']

# getting dataframe after removing the columns
df1=df.drop(columns=columns_to_drop)
print(df1.shape)
df1
```

```python
df1.columns
df1.describe()
df1.fillna(0)
```

```python
# selecting our target variables to give to the models
y1=df1['seasonal_vaccine']
y2=df1['xyz_vaccine']
X=df1.drop(columns=['seasonal_vaccine','respondent_id','xyz_vaccine'])
X=X.fillna(0)
```

**Model_1== Gaussian Niave Bayes**

```python
# now fitting this trainig data to naive bayes model
 # now fitting this trainig data to naive bayes model for the getting first target variable prediction

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model1=GaussianNB()

model1.fit(X,y1)

```

```python
 # now predicting on the test data for first target variable
# loading test csv
df_test=pd.read_csv(r'/content/test_set_features.csv')
df_test.drop(columns=columns_to_drop,inplace=True)
df_test.fillna(0,inplace=True)
X_test=df_test.drop(columns=['respondent_id'])
#y_pred=model.predict(X_test) # for getting prediction in binary form
# to get prediction in probability
y_pred1=model1.predict_proba(X_test)
```

```python
arr1 = np.array(y_pred1)
arr1=np.max(arr1,axis=1)
print(arr1)
seasonal_vaccine_pred=arr1
```

```python
 # now fitting this trainig data to naive bayes model for the getting second target variable prediction

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model2=GaussianNB()

model2.fit(X,y2)

```

```python
 # now predicting on the test data for second target variable
 y_pred2=model2.predict_proba(X_test)
 arr2=np.array(y_pred2)
 arr2=np.max(arr2,axis=1)
 print(arr2)
 xyz_vaccine_pred=arr2
 xyz_vaccine_pred.shape
```

```python
# now we have to return the csv file output as given in the assignment
# for that we need to combine three columns respondent_id, seasonal_vaccine_pred=arr1,  xyz_vaccine_pred
df_output=pd.DataFrame(list(zip(df_test['respondent_id'],xyz_vaccine_pred,seasonal_vaccine_pred)),columns=['respondent_id','xyz_vaccine','seasonal_vaccine'])
df_output
# now we will downloading the df_output csv file
df_output.to_csv('output_using_Niave_Bayes.csv',index=False)
df_output
```
